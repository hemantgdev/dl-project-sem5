```python
import re
import csv

# Specify input and output file paths here
INPUT_FILE = '/Users/hemantg/Desktop/ipl2011.txt'
OUTPUT_FILE = '/Users/hemantg/Desktop/espn_cricket_links_2011.csv'

def extract_story_links(input_file, output_file):
    # Base URL for ESPN Cricket
    base_url = 'https://www.espncricinfo.com'
    
    # Store unique links to avoid duplicates
    unique_links = set()
    
    # Read input text file
    with open(input_file, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # Find all /story links using regex
    story_links = re.findall(r'(/story[^\s"\']+)', content)
    
    # Add base URL and remove duplicates
    full_links = [base_url + link for link in story_links]
    unique_links.update(full_links)
    
    # Write links to CSV
    with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Link'])  # Header
        for link in unique_links:
            writer.writerow([link])
    
    print(f'Extracted {len(unique_links)} unique story links to {output_file}')

# Run the extraction
extract_story_links(INPUT_FILE, OUTPUT_FILE)
```

```python
import pandas as pd
import requests

# ========== CONFIGURATION SECTION ==========
# SPECIFY YOUR INPUT AND OUTPUT CSV PATHS HERE
INPUT_CSV_PATH = '/Users/hemantg/Desktop/match-rp-scraping.csv'  # Replace with your input CSV path
OUTPUT_CSV_PATH = '/Users/hemantg/Desktop/output_rp_links.csv'  # Replace with your desired output CSV path
# ==========================================

def generate_match_links(input_csv_path, output_csv_path):
    """
    Generate ESPN CricInfo match links for a specific IPL series
    
    Parameters:
    - input_csv_path: Path to the input CSV file with match details
    - output_csv_path: Path to save the output CSV with generated links
    """
    # Constant base URL for the 2007-08 IPL series
    base_series_url = "https://www.espncricinfo.com/series/indian-premier-league-2007-08-313494/"
    
    # Read the input CSV
    df = pd.read_csv(input_csv_path)
    
    # Prepare lists to store results
    match_links = []
    processed_ids = []
    
    # Start with the first match ID
    current_match_id = df.iloc[0]['ID']
    current_match_number = 1
    
    while True:
        try:
            # Find the row corresponding to the current match ID
            match_row = df[df['ID'] == current_match_id].iloc[0]
            
            # Prepare URL components
            # Convert team names to lowercase with hyphens
            team1_slug = match_row['Team1'].lower().replace(' ', '-')
            team2_slug = match_row['Team2'].lower().replace(' ', '-')
            
            # Construct full match URL
            full_match_url = f"{base_series_url}{team1_slug}-vs-{team2_slug}-{current_match_number}rd-match-{current_match_id}"
            
            # Validate the URL
            response = requests.head(full_match_url)
            
            # If URL is valid, add to results
            if response.status_code == 200:
                match_links.append(full_match_url)
                processed_ids.append(current_match_id)
                current_match_number += 1
                
                # Move to next unique match ID
                next_match_indices = df[df['ID'] > current_match_id].index
                if len(next_match_indices) > 0:
                    current_match_id = df.loc[next_match_indices[0], 'ID']
                else:
                    break
            else:
                # Try alternative match numbering (1st, 2nd, 3rd, etc.)
                ordinal_urls = [
                    f"{base_series_url}{team1_slug}-vs-{team2_slug}-{current_match_number}st-match-{current_match_id}",
                    f"{base_series_url}{team1_slug}-vs-{team2_slug}-{current_match_number}nd-match-{current_match_id}",
                    f"{base_series_url}{team1_slug}-vs-{team2_slug}-{current_match_number}th-match-{current_match_id}"
                ]
                
                url_found = False
                for alt_url in ordinal_urls:
                    response = requests.head(alt_url)
                    if response.status_code == 200:
                        match_links.append(alt_url)
                        processed_ids.append(current_match_id)
                        current_match_number += 1
                        
                        # Move to next unique match ID
                        next_match_indices = df[df['ID'] > current_match_id].index
                        if len(next_match_indices) > 0:
                            current_match_id = df.loc[next_match_indices[0], 'ID']
                        else:
                            break
                        
                        url_found = True
                        break
                
                # If no valid URL found, break the loop
                if not url_found:
                    break
        
        except IndexError:
            # If no matching row is found, break the loop
            break
        except Exception as e:
            print(f"Error processing match {current_match_id}: {e}")
            break
    
    # Create a DataFrame of the results
    results_df = pd.DataFrame({
        'MatchID': processed_ids,
        'MatchLink': match_links
    })
    
    # Save to CSV
    results_df.to_csv(output_csv_path, index=False)
    
    print(f"Generated {len(results_df)} match links. Saved to {output_csv_path}")
    
    return results_df

# Run the script
def main():
    generate_match_links(INPUT_CSV_PATH, OUTPUT_CSV_PATH)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import requests

# ========== CONFIGURATION SECTION ==========
INPUT_CSV_PATH = '/Users/hemantg/Desktop/match-rp-scraping.csv'  # Replace with your input CSV path
OUTPUT_CSV_PATH = '/Users/hemantg/Desktop/output_rp_links3.csv'  # Replace with your desired output CSV path
# ==========================================

def generate_match_links(input_csv_path, output_csv_path):
    """
    Generate ESPN CricInfo match links for a specific IPL series
    
    Parameters:
    - input_csv_path: Path to the input CSV file with match details
    - output_csv_path: Path to save the output CSV with generated links
    """
    # Constant base URL for the 2007-08 IPL series
    base_series_url = "https://www.espncricinfo.com/series/indian-premier-league-2007-08-313494/"
    
    # Read the input CSV
    df = pd.read_csv(input_csv_path)
    
    # Prepare lists to store results
    match_links = []
    
    print(f"Processing {len(df)} matches...")
    
    # Iterate through all matches in the DataFrame
    for index, match_row in df.iterrows():
        current_match_id = match_row['ID']
        match_number = match_row['match no.'].strip().lower()  # Match number (e.g., '1st', '2nd', etc.)
        team1_slug = match_row['Team1'].lower().replace(' ', '-')
        team2_slug = match_row['Team2'].lower().replace(' ', '-')
        
        base_match_url = f"{base_series_url}{team1_slug}-vs-{team2_slug}-{match_number}-match-{current_match_id}"
        preview_url = f"{base_match_url}/match-preview"
        report_url = f"{base_match_url}/match-report"
        
        # Log generated URLs
        print(f"Generated URLs for Match ID {current_match_id}:")
        print(f"  Preview URL: {preview_url}")
        print(f"  Report URL: {report_url}")
        
        # Initialize list to hold valid links for this match
        valid_links = []
        
        # Check URLs
        try:
            # Validate preview URL
            preview_response = requests.head(preview_url, timeout=10)
            if preview_response.status_code == 200:
                valid_links.append(preview_url)
            else:
                print(f"Preview link invalid for Match ID {current_match_id}.")
            
            # Validate report URL
            report_response = requests.head(report_url, timeout=10)
            if report_response.status_code == 200:
                valid_links.append(report_url)
            else:
                print(f"Report link invalid for Match ID {current_match_id}.")
            
            # If no valid links, log an error
            if not valid_links:
                print(f"Error: No valid links found for Match ID {current_match_id}.")
            else:
                match_links.append({
                    'match_id': current_match_id,
                    'links': ' | '.join(valid_links)
                })
        
        except requests.RequestException as e:
            print(f"Request error for Match ID {current_match_id}: {e}")
            continue
        except Exception as e:
            print(f"Unexpected error for Match ID {current_match_id}: {e}")
            continue
    
    # Create a DataFrame of the results
    results_df = pd.DataFrame(match_links)
    
    # Save to CSV
    results_df.to_csv(output_csv_path, index=False)
    
    print(f"Generated {len(results_df)} valid match links out of {len(df)} matches. Saved to {output_csv_path}")
    
    return results_df

# Run the script
def main():
    generate_match_links(INPUT_CSV_PATH, OUTPUT_CSV_PATH)

if __name__ == "__main__":
    main()
```

```python
import re
import csv

# Read the file content
file_path = '/Users/hemantg/Desktop/ipl2008-matches.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# Regular expression to find all href links containing "/series"
pattern = r'href="(/series[^"]+)"'
matches = re.findall(pattern, content)

# Prepend the base URL
base_url = "https://www.espncricinfo.com"
full_links = [base_url + link for link in matches]

# Remove every alternate link
valuable_links = [link for i, link in enumerate(full_links) if i % 2 == 0]

# Modify links for match-preview and match-report
final_links = []
for link in valuable_links:
    if "full-scorecard" in link:
        match_preview = link.replace("full-scorecard", "match-preview")
        match_report = link.replace("full-scorecard", "match-report")
        final_links.extend([match_preview, match_report])
    else:
        final_links.append(link)

# Write the links to a CSV file
output_csv = '/Users/hemantg/Desktop/ipl2008-matches-rp1.csv'
with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["Links"])
    for link in final_links:
        csv_writer.writerow([link])

print(f"Processed links and saved them to {output_csv}.")
```

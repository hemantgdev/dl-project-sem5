```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import numpy as np

# Initialize the embedding model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# List of names
names_list = df1.player.unique().tolist()

# Generate embeddings for the list of names
name_embeddings = model.encode(names_list, convert_to_tensor=True)


# Specify the column name containing names in the CSV
column_name = 'player'  # Replace with the actual column name
csv_names = df2[column_name].tolist()

# Generate embeddings for the names in the CSV column
csv_embeddings = model.encode(csv_names, convert_to_tensor=True)

# Find closest matches
matches = []
for i, name in enumerate(names_list):
    # Compute cosine similarities
    similarities = util.cos_sim(name_embeddings[i], csv_embeddings).numpy().flatten()
    best_match_idx = np.argmax(similarities)
    matches.append({
        "Name": name,
        "Closest Match": csv_names[best_match_idx],
        "Similarity Score": similarities[best_match_idx]
    })

# Convert matches to a DataFrame and display
matches_df = pd.DataFrame(matches)
print(matches_df)

# Save results to a CSV file (optional)
output_file = 'name_matches.csv'
matches_df.to_csv(output_file, index=False)
print(f"Results saved to {output_file}")
```
